```python
import ast
l  = "[1,2,3,4,5]"
ast.literal_eval(l)
```

```python
ast.literal_eval(df['belongs_to_collection'].values[0])['name']
```

```python
# store the values in a new list
sha = []
for i in df['belongs_to_collection']:
  if i == i:
    sha.append(ast.literal_eval(i)['name'])
  else:
    sha.append("no collection")
```

```python
# create a new column called collections
df['collection_name'] = sha
```
